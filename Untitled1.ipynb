```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Walmart.csv')

# Convert 'Date' to datetime format if not already
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')

# Display basic information about the dataset
print("Dataset Info:")
print(data.info())
print("\nFirst Few Rows:")
print(data.head())
print("\nSummary Statistics:")
print(data.describe())

# 1. Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# 2. Distribution of Weekly Sales
plt.figure(figsize=(10, 6))
sns.histplot(data['Weekly_Sales'], kde=True, color='skyblue')
plt.title('Distribution of Weekly Sales')
plt.xlabel('Weekly Sales')
plt.ylabel('Frequency')
plt.show()

# 5. Correlation Heatmap
plt.figure(figsize=(10, 6))
correlation = data[['Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Temperature', y='Weekly_Sales', hue='Holiday_Flag', palette='coolwarm')
plt.title('Weekly Sales vs. Temperature')
plt.xlabel('Temperature (F)')
plt.ylabel('Weekly Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Fuel_Price', y='Weekly_Sales', hue='Holiday_Flag', palette='viridis')
plt.title('Weekly Sales vs. Fuel Price')
plt.xlabel('Fuel Price')
plt.ylabel('Weekly Sales')
plt.show()
```

```python
data['Month'] = data['Date'].dt.month
monthly_avg_sales = data.groupby('Month')['Weekly_Sales'].mean()

plt.figure(figsize=(10, 6))
monthly_avg_sales.plot(kind='line', color='skyblue', marker='o')
plt.title('Average Weekly Sales by Month')
plt.xlabel('Month')
plt.ylabel('Average Weekly Sales')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

```python
store_avg_sales = data.groupby('Store')['Weekly_Sales'].mean().sort_values()

plt.figure(figsize=(12, 6))
store_avg_sales.plot(kind='bar', color='lightgreen')
plt.title('Average Weekly Sales by Store')
plt.xlabel('Store')
plt.ylabel('Average Weekly Sales')
plt.show()
```

```python
data['Day_of_Week'] = data['Date'].dt.day_name()  # Extract day of the week
day_avg_sales = data.groupby('Day_of_Week')['Weekly_Sales'].mean().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

plt.figure(figsize=(10, 6))
day_avg_sales.plot(kind='bar', color='cornflowerblue')
plt.title('Average Weekly Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Weekly Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
data['Unemployment_Bin'] = pd.cut(data['Unemployment'], bins=[4, 5, 6, 7, 8, 9, 10],
                                  labels=['4-5%', '5-6%', '6-7%', '7-8%', '8-9%', '9-10%'])
unemployment_avg_sales = data.groupby('Unemployment_Bin')['Weekly_Sales'].mean()

plt.figure(figsize=(10, 6))
unemployment_avg_sales.plot(kind='bar', color='steelblue')
plt.title('Average Weekly Sales by Unemployment Rate')
plt.xlabel('Unemployment Rate Bin')
plt.ylabel('Average Weekly Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Weekly_Sales'], bins=30, color='teal', edgecolor='black')
plt.title('Distribution of Weekly Sales')
plt.xlabel('Weekly Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
from pandas.plotting import lag_plot
plt.figure(figsize=(6, 6))
lag_plot(data['Weekly_Sales'], lag=1)
plt.title('Lag Plot of Weekly Sales')
plt.xlabel('Weekly Sales (t)')
plt.ylabel('Weekly Sales (t + 1)')
plt.show()
```
